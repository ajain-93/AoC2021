```
string filename = @"C:\Temp\Advent of Code 2021\Day16.txt";
string[] raw_lines = System.IO.File.ReadAllLines(filename);
```

```

string hexValue = raw_lines[0];
Console.WriteLine($"{hexValue.Length} {hexValue}");
string binValueMain = "";
foreach (var c in hexValue){
    int hex2dec = int.Parse(c.ToString(), System.Globalization.NumberStyles.HexNumber);
    string charBin = Convert.ToString(hex2dec,2).PadLeft(4,'0');
    // Console.WriteLine($"{c} {hex2dec} {charBin}");
    binValueMain = $"{binValueMain}{charBin}";
}
// Console.WriteLine($"Package: {binValue.Length} {binValue}");
// return;

//  Convert.ToString(Convert.ToInt64(hexValue, 16), 2).PadLeft(4 * hexValue.Length, '0');

Int64 TypeOperation(int type, Int64 value1, Int64 value2){
    Console.WriteLine($"Operation {type} 1:{value1} 2:{value2}");
    switch (type){
        case 0:
            return value1 + value2;
        case 1:
            return value1 * value2;
        case 2:
            return Math.Min(value1,value2);
        case 3:
            return Math.Max(value1,value2);
        case 5:
            return (value1 > value2) ? 1 : 0;
        case 6:
            return (value1 < value2) ? 1 : 0;
        case 7:
            return (value1 == value2) ? 1 : 0;
    }
    return -1;
}


int overalli = 0;
int sumVersions = 0;
Int64 interpretPackage(string binValue) {
    // Console.WriteLine($"Package: {binValue.Length} {binValue}");

    string versionBin = binValue.Substring(0,3);
    int versionDec = Convert.ToInt32(versionBin, 2);
    sumVersions += versionDec;
    // Console.WriteLine($"Version: {versionDec} ({versionBin})");

    string typeBin = binValue.Substring(3,3);
    int typeDec = Convert.ToInt32(typeBin, 2);
    Console.WriteLine($"Type: {typeDec} ({typeBin})");

    switch (typeDec) {
        case 4:
            //Literal value
            int i=6;
            overalli += i;
            string valueBin = "";
                while(true){
                    overalli +=5;
                    valueBin = valueBin + binValue.Substring(i+1,4);                
                    if (binValue.Substring(i,1) == "0"){
                        Int64 valueDec = Convert.ToInt64(valueBin, 2);
                        Console.WriteLine($"Literal: {valueDec} ({valueBin})");    
                        return valueDec;
                    } else {
                        i+=5;
                    }
                }
        default:
            // Operator Packet
            // Console.WriteLine($"Length Type: {binValue.Substring(6,1)}");

            if (binValue.Substring(6,1) == "0") {
                string lenIDbin = binValue.Substring(7,15);
                int lenId = Convert.ToInt32(lenIDbin, 2);
                Console.WriteLine($"Length: {lenId} ({lenIDbin})");
                overalli = overalli + 7+15;
                int max = overalli + lenId;
                Int64 val = interpretPackage(binValueMain.Substring(overalli));
                while (overalli< max) {
                    Console.WriteLine($"i: {overalli}");
                    val = TypeOperation(typeDec,val, interpretPackage(binValueMain.Substring(overalli)));
                }
                return val;
            } else {
                string lenIDbin = binValue.Substring(7,11);
                int lenId = Convert.ToInt32(lenIDbin, 2);
                Console.WriteLine($"Count: {lenId} ({lenIDbin})");
                overalli = overalli + 7+11;
                Int64 val = interpretPackage(binValueMain.Substring(overalli));
                for (int j=1;j<lenId;j++) {
                    val = TypeOperation(typeDec,val, interpretPackage(binValueMain.Substring(overalli)));
                }
                return val;
            }
    }
    return -1;
}

Console.WriteLine($"\n\n16B: {interpretPackage(binValueMain)}");
Console.WriteLine($"16A: {sumVersions}");
```
