```
string filename = @"C:\Temp\Advent of Code 2021\Day15.txt";
string[] raw_lines = System.IO.File.ReadAllLines(filename);
```

```

int max_x = raw_lines[0].Count();
int max_y = raw_lines.Count();

int[,] board = new int[max_x, max_y];
bool[,] visited = new bool[max_x, max_y];
int[,] pathWeight = new int[max_x, max_y];

for(int x=0;x<max_x;x++){
    for(int y=0;y<max_y;y++){
        board[x,y] = int.Parse(raw_lines[y][x].ToString());
        pathWeight[x,y] = int.MaxValue;
    }
}

void Print(int[,] board) {
    for(int y=0;y<board.GetLength(1);y++){
        for(int x=0;x<board.GetLength(0);x++){
            string val = (board[x,y] == int.MaxValue) ? "  " : String.Format("{0,1}",board[x,y]);
            Console.Write($"{val} ");
        }
        Console.WriteLine();
    }
    Console.WriteLine();
}
void PrintVisited(bool[,] board) {
    for(int y=0;y<board.GetLength(1);y++){
        for(int x=0;x<board.GetLength(0);x++){
            Console.Write($"{board[x,y]} ");
        }
        Console.WriteLine();
    }
    Console.WriteLine();
}

void step(int x, int y){
    if (visited[x,y]) {
        return;
    } else {
        visited[x,y] = true;
    }
    int current = pathWeight[x,y];

    
    bool isNotLeftBorder = x > 0;
    bool isNotRightBorder = x < board.GetLength(0)-1;
    bool isNotTopBorder = y > 0;
    bool isNotBottomBorder = y < board.GetLength(1)-1;

    int up = int.MaxValue;
    int down = int.MaxValue;
    int left = int.MaxValue;
    int right = int.MaxValue;

    if (isNotRightBorder){
        if (pathWeight[x+1,y] >= current + board[x+1,y]) {
            pathWeight[x+1,y] = current + board[x+1,y];
        }
        right = pathWeight[x+1,y];
    }
    if (isNotLeftBorder){
        if (pathWeight[x-1,y] >= current + board[x-1,y]) {
            pathWeight[x-1,y] = current + board[x-1,y];
        }
        left = pathWeight[x-1,y];
    }   
    if (isNotTopBorder){
        if (pathWeight[x,y-1] >= current + board[x,y-1]) {
            pathWeight[x,y-1] = current + board[x,y-1];
        }
        up = pathWeight[x,y-1];
    }
    if (isNotBottomBorder){
        if (pathWeight[x,y+1] >= current + board[x,y+1]) {
            pathWeight[x,y+1] = current + board[x,y+1];
        }
        down = pathWeight[x,y+1];
    }
}

int minValue = int.MaxValue;
int next_x = 0;
int next_y = 0;
Console.WriteLine($"Starting Values: {next_x}, {next_y}");


void nextStep() {
    minValue = int.MaxValue;
    for(int x=0;x<max_x;x++){
        for(int y=0;y<max_y;y++){
            if (visited[x,y]){
                // Console.WriteLine($"Skip {x},{y} pathWeight[x,y]");
            } else {
            int currentVal = pathWeight[x,y];
                if (currentVal < minValue){
                    minValue = currentVal;
                    next_x = x;
                    next_y = y;
                    // Console.WriteLine($"Min: {x},{y} {currentVal}  visited[x,y]");
                }
            }
        }
    }
}


pathWeight[0,0] = 0;
// int steps = 100;
// for (int i=0;i<steps;i++) {
while (true) {
    step(next_x,next_y);
    nextStep();    
    if (next_x == max_x-1 && next_y == max_y-1){
        Console.WriteLine($"Next: {next_x},{next_y}: {minValue}");
        break;
    }
}

// Console.WriteLine($"\nBoards");
// Print(board);
// Print(pathWeight);
// PrintVisited(visited);
```
